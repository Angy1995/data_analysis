# VLOOKUP

```python
# transactions
left_df = pd.DataFrame({'transaction_id': ['A', 'B', 'C', 'D'], 
                       'user_id': ['Peter', 'John', 'John', 'Anna'],
                       'value': np.random.randn(4),
                      })
# users
right_df = pd.DataFrame({'user_id': ['Paul', 'Mary', 'John',
                                     'Anna'],
                        'favorite_color': ['blue', 'blue', 'red', 
                                           np.NaN],
                       })
```

```python
left_df
```

```python
right_df
```

```python
left_df.merge(right_df, on='user_id', how='left')
```

# SUMIF 和数据透视

```python
# 分组统计均值、方差等

import pandas as pd

df = pd.read_csv(CSV_FILE_PATH)

def period(age):
    if 0 < age < 12:
        return 'Child'
    elif age < 18:
        return 'Teenager'
    else:
        return 'Adult'
    
df['if_adult'] = df['Age'].apply(period)

df.groupby(['if_adult']).sum()
```

```python
# 数据透视表

pd.pivot_table(df, values='Survived', index=['Pclass', 'Sex'], columns=['if_adult'])
```
